```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("Samsung_mobile - Copy.csv")

# Drop any duplicates
df.drop_duplicates(inplace=True)

# Handle missing values if any
df.dropna(inplace=True)  # Or use df.fillna() to fill with specific values

# Remove columns that are not needed
df = df.drop(["Additional_Features"], axis=1)

# Convert data types if necessary
df['Ram_GB'] = df['Ram_GB'].astype(int)
df['Rom_GB'] = df['Rom_GB'].astype(int)
df['Battery_mAh'] = df['Battery_mAh'].astype(int)
df['Latest_Price_Rupees'] = df['Latest_Price_Rupees'].astype(int)
df['Possible_Discount_Rupees'] = df['Possible_Discount_Rupees'].astype(int)
df['Final_Price_Rupees'] = df['Final_Price_Rupees'].astype(int)
df['Reviews (out of 5)'] = df['Reviews (out of 5)'].astype(float)

print("Data Cleaning Done")
print(df.head())
```

```python
# Summary statistics
print(df.describe())

# Distribution of Reviews
plt.figure(figsize=(10, 6))
df['Reviews (out of 5)'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Review Ratings')
plt.ylabel('Number of Phones')
plt.title('Distribution of Samsung Phone Reviews')
plt.show()

# Average Price vs Reviews
avg_price = df.groupby('Reviews (out of 5)')['Final_Price_Rupees'].mean()
plt.figure(figsize=(10, 6))
avg_price.plot(kind='bar')
plt.xlabel('Review Ratings')
plt.ylabel('Average Final Price (Rupees)')
plt.title('Average Final Price vs. Reviews')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Final_Price_Rupees'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Final Price (Rupees)')
plt.ylabel('Number of Phones')
plt.title('Distribution of Final Prices of Samsung Phones')
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(8, 8))
df['Reviews (out of 5)'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of Review Ratings')
plt.ylabel('')  # Hide y-label
plt.show()
```

```python
# Line Plot
avg_price = df.groupby('Reviews (out of 5)')['Final_Price_Rupees'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_price.index, avg_price.values, marker='o', linestyle='-', color='b')
plt.xlabel('Review Ratings')
plt.ylabel('Average Final Price (Rupees)')
plt.title('Average Final Price vs. Review Ratings')
plt.grid(True)
plt.show()
```
